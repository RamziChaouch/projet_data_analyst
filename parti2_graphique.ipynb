### J'ai essayez de créer un graphique du nombre total de vaccinations par pays, en utilisant les données d'un DataFrame Pandas appelé "df".

# Pour créer ce graphique, j'ai utilise la méthode groupby() du DataFrame pour regrouper les données par pays, puis j'ai utiliser la méthode sum() pour calculer le nombre total de vaccinations pour chaque pays.et j'ai utilise la méthode plot() pour créer un graphique linéaire du total des vaccinations par pays.

```python
# J'ai essayez de créer un nuage de points 3D du total des vaccinations par rapport au pays, en utilisant les données
#d'un Pandas DataFrame appelé "df".
#Pour créer ce tracé, j'ai utilise la fonction scatter3D() du module axes3d de Matplotlib, qui nous permet de spécifier
#les données des axes x, y et z pour le tracé. Dans ce cas, l'axe des x sera la colonne "total_vaccinations"
#et l'axe des y sera la colonne "pays".
#Avant de pouvoir créer le nuage de points, j'ai utiliser la classe LabelEncoder de scikit-learn pour encoder
#la colonne "pays" afin qu'elle puisse être utilisée comme données de l'axe y. La classe LabelEncoder attribuera
#une valeur entière unique à chaque valeur unique dans la colonne "pays", qui peut ensuite être utilisée comme
#données d'axe y pour le tracé.
```

```python
import matplotlib.pyplot as plt
x1 = df['total_vaccinations']
y1 = df['country']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y1  = encoder.fit_transform(y1)
ax = plt.axes(projection='3d')
ax.scatter3D(x1,y1)
```

```python
# J'ai essayez de créer une grille de tracés 2x2, chaque tracé affichant une colonne de données différente à partir
#d'un DataFrame Pandas appelé "df". Les données du DataFrame sont filtrées pour n'inclure que les lignes pour
#lesquelles la colonne "iso_code" est égale à "TUN".
#Pour créer la grille de parcelles,j'ai utilise la fonction subplots() de Matplotlib pour créer une figure
#avec une grille 2x2 de sous-parcelles, puis j'ai utilise la méthode plot() de chaque sous-parcelle pour tracer les données.
#j'ai utiliser la méthode set_title() pour définir le titre de chaque tracé.
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,p1 =  plt.subplots(2,2)
plt.subplots_adjust(wspace=0.4,hspace=0.4)
subset = df[(df.iso_code =="TUN")]
p1[0][0].plot(subset['total_vaccinations'])
p1[0][0].set_title('total')


p1[0][1].plot(subset['vaccines'])
p1[0][1].set_title('nom vaccin')


p1[1][0].plot(subset['people_vaccinated'])
p1[1][0].set_title('people vaccinee')


p1[1][1].plot(subset['country'])
p1[1][1].set_title('nom country')
plt.show()
```
