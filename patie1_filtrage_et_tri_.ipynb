### L'indexeur df.iloc peut être utilisé pour sélectionner des lignes et des colonnes à partir d'un DataFrame en fonction de leur position entière. Dans ce cas, l'expression df.iloc[-5:,:] sélectionnera les cinq dernières lignes du DataFrame et toutes les colonnes.
# La partie -5: de l'indexeur spécifie les lignes à sélectionner, et le : après la virgule spécifie que toutes les colonnes doivent être sélectionnées. L'index négatif dans l'indexeur de lignes est utilisé pour sélectionner les lignes à partir de la fin du DataFrame, -1 étant la dernière ligne, -2 l'avant-dernière, et ainsi de suite.

```python
df.iloc[-5:,:]
```

```python
#5 premières lignes et colonnes 0, 2 et 3
#on a une liste d'indices en colonne
```

### L'indexeur df.iloc peut être utilisé pour sélectionner des lignes et des colonnes à partir d'un DataFrame en fonction de leur position entière. Dans ce cas, l'expression df.iloc[0:5,[0,2,3]] sélectionnera les cinq premières lignes (lignes 0 à 4) et les colonnes aux positions 0, 2 et 3.
# La partie 0:5 de l'indexeur spécifie les lignes à sélectionner et la partie [0,2,3] spécifie les colonnes à sélectionner. Les deux-points (:) dans l'indexeur de lignes sont utilisés pour spécifier une plage de lignes, et la liste d'entiers dans l'indexeur de colonnes est utilisée pour spécifier les colonnes spécifiques à sélectionner.

```python
df.iloc[0:5,[0,2,3]]
```
